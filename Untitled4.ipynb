```python
#GDP FORECAST WITH DUMMY VALUES
# ðŸ“¦ Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# ðŸ“Š Step 1: Create GDP data (in trillion USD)
data = {
    'Year': [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
    'GDP': [1.8, 1.92, 2.03, 2.1, 2.28, 2.42, 2.53, 2.3, 2.55, 2.7]
}

df = pd.DataFrame(data)

# ðŸ§® Step 2: Prepare data for model
X = df[['Year']]              # independent variable must be 2D
y = df['GDP']                 # dependent variable

# Create and train model
model = LinearRegression()
model.fit(X, y)

# ðŸ§  Step 3: Predict GDP for future year(s)
future_years = np.array([[2023], [2024], [2025]])   # predict next 3 years
predicted_gdp = model.predict(future_years)

# Combine future predictions into a new DataFrame
forecast_df = pd.DataFrame({
    'Year': future_years.flatten(),
    'Predicted GDP': predicted_gdp
})

# Display results
print("ðŸ“ˆ Future GDP Forecast:")
print(forecast_df)

# ðŸŽ¨ Step 4: Visualize results
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['GDP'], 'bo-', label='Actual GDP')
plt.plot(forecast_df['Year'], forecast_df['Predicted GDP'], 'r--', label='Forecasted GDP')
plt.title('GDP Growth Forecast using Linear Regression')
plt.xlabel('Year')
plt.ylabel('GDP (Trillion USD)')
plt.legend()
plt.grid(True)
plt.show()

```
