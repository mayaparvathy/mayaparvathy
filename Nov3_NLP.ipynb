#TASK 1

Read the sentences and label them manually as: Positive / Negative / Neutral


```python
#process 1
import pandas as pd

data = {
    "text": [
        "I'm tired today",
        "This drink is cool",
        "I'm going home today",
        "I hate homework",
        "The weather is sunny",
        "Oh great,it's clumsy",
        "Oh great,it's raining"
    ]
}

df = pd.DataFrame(data)
df
```

```python
#Process 2
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

Change the sentences in process 2

```python
#process 3
!pip install afinn
from afinn import Afinn

af = Afinn()

sentences = ["I hate walking", "I'm going home", "This is fine"]
for s in sentences:
    print(s, af.score(s))
```

```python
#process 4
!pip install textblob
from textblob import TextBlob

sentences = ["I hate walking", "I love eating", "This is fine"]
for s in sentences:
    blob = TextBlob(s)
    print(s, blob.sentiment)
```

how VADER differs form Affin and TextBlob?

Affin's output is a numeric score while TextBlob's output includes both polarity and subjectivity.

```python
#Let us compare !
from textblob import TextBlob

df['textblob_score'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['textblob_sentiment'] = df['textblob_score'].apply(
    lambda x: 'Positive' if x>0 else ('Negative' if x<0 else 'Neutral')
)
df
```

```python
###TASK 2
import pandas as pd

data = {
    "text": [
         "I HATE this movie!!! ",
    "This course is SO Good...  ",
    "I think AI is exploitative",
    "Not sure if I like this. ",
    "Great work!!! Keep it up "
    ]
}

df = pd.DataFrame(data)
df
```

Put it in dataframe

```python
df['clean_text'] = df['text'].str.lower()
df['clean_text'] = df['clean_text'].str.strip()
df[['text','clean_text']]
```

Why do we convert text to lowercase?


To maintain uniformity and improve accuracy.

```python
import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]
```

What happens to meaning if punctuation is removed?

It affects a sentence by changing its actual meaning.

```python

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]
```

Why remove stopwords?


To make analysis more efficient and meaningful.

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]
```

Can emojis express emotion? Give an example.

Yes, Emojis express emotions. In the above result, Hate movie with emojis and without emojis emote two different feelings.
